```python
## Extract TTV and get counts
list_ttv = ['NC_014073', 'NC_014081', 'NC_014084', 'NC_014080', 'NC_014075', 'NC_014094', 'NC_014074',
            'NC_014096', 'NC_014078', 'NC_014079', 'NC_026664', 'NC_014091', 'NC_014077', 'NC_014076',
            'NC_026765', 'NC_026764', 'NC_026662', 'NC_014069', 'NC_014097', 'NC_014088', 'NC_025727',
            'NC_014082', 'NC_020498', 'NC_014089', 'NC_025726', 'NC_014095', 'NC_014086', 'NC_015783']
ttv_df = da[list_ttv + ["sample"]].copy()
ttv_df["counts_TTV"] = ttv_df[list_ttv].sum(axis=1)
ttv_df.head()
```

```python
fig = plt.figure(figsize = (40, 10))
ax = fig.add_subplot(111)
ax.plot(ttv_df)
plt.xticks(rotation=90)
plt.show()
```

```python
# Extract TTMV and get counts
list_ttmv = ['NC_030297', 'NC_014068', 'NC_014090']
ttmv_df = da[list_ttmv + ["sample"]].copy()
ttmv_df["counts_TTMV"] = da[list_ttmv].sum(axis=1)
```

```python
# Extract TTMDV and get counts
list_ttmDv = ['NC_014093']
ttmDv_df = da[list_ttmDv + ["sample"]].copy()
ttmDv_df['counts_TTMDV'] = da["NC_014093"]
```

```python
## Percentajes per virus type and sample

totals = [i+j+k for i,j,k in zip(new_df["TTV"], 
                                 new_df["TTMV"], 
                                new_df["TTMDV"])]
for n, i in enumerate(totals):
    if i == 0:
        totals[n] = 1

TTV = [i/j*100 for i, j in zip(new_df["TTV"], totals)]
TTMV = [i/j*100 for i, j in zip(new_df["TTMV"], totals)]
TTMDV = [i/j*100 for i, j in zip(new_df["TTMDV"], totals)]

# plot
fig = plt.figure(figsize = (40, 10))
barWidth=0.8
names = ('11-2', '13-1', '13-2', '16-1', '16-3', '17-1', '17-2', '19-1', '19-2', 
         '21-1', '41-2', '43-1', '43-2', '51-1', '52-2', '53-1', '53-2', '56-1', '56-2', 
         '58-1', '61-1', '61-2', '7-2', '9-1', '9-2', 'JFELIP-1', 'JFELIP-2', 'MAICAS-1', 
         'MATEU-1', 'MATEU-2', 'PAC13-1', 'PAC13-2', 'PAC20-1', 'PAC20-2', 'PAC2-2', 
         'PAC3-1', 'PAC3-2', 'PAC41-1', 'PONS1-1', 'SANTOS-1', 'SANTOS-2', 'TTV1', 
         'TTV2')
r = []
for i in range(1, 44):
    r.append(i)
    
plt.bar(r, TTV, color='#FFA500', edgecolor='white', width=barWidth, label="TTV")
plt.bar(r, TTMV, bottom=TTV, color='#FF0000', 
       edgecolor='white', width=barWidth, label="TTMV")
plt.bar(r, TTMDV, bottom=[i+j for i, j in zip(TTV,
        TTMV)], color='#0000FF', edgecolor='white', width=barWidth, label="TTMDV")

plt.xticks(r, names, rotation=60,fontsize=16)
plt.yticks(fontsize=20)
plt.legend(loc='upper left', bbox_to_anchor=(1,1), ncol=1, fontsize=20)
plt.autoscale()
plt.savefig("stacked_bars-mar2019.png")
plt.show()
```
