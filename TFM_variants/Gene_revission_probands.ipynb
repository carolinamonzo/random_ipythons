```python
import os
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
plt.style.use('seaborn-colorblind')
sns.set_style('darkgrid')
```

```python
## DE NOVO

# Read file with de_novo genes
with open("./de_novo/de_novo_families.tsv", "r") as fi:
    de_novo = "\t".join(fi.read().splitlines()).split("\t")

# Filter out empty string elements in the list
de_novoF = list(filter(None, de_novo))

# Get number of appearances of genes 
de_novo_keys = list(set(de_novoF))

counts = []

for i in range(0, len(de_novo_keys)):
    counts.append(de_novoF.count(de_novo_keys[i]))

# Create a dictionary with name of gene and appearances in the prioritization
dict_de_novo = dict(zip(de_novo_keys, counts))
dict_de_novo
```

```python
## MENDEL ERRORS

# Read file with mendel errors
with open("./mendel_errors/mendel_errors_families.tsv", "r") as fi:
    mendel = "\t".join(fi.read().splitlines()).split("\t")

# Filter out empty string elements in the list
mendelF = list(filter(None, mendel))

# Get number of appearances of genes 
mendel_keys = list(set(mendelF))

counts = []

for i in range(0, len(mendel_keys)):
    counts.append(mendelF.count(mendel_keys[i]))

# Create a dictionary with name of gene and appearances in the prioritization
dict_mendel_error = dict(zip(mendel_keys, counts))
dict_mendel_error
```

```python
## AUTOSOMAL RECESSIVE WITH HIGH IMPACT

# Read file with autosomal_recessive
with open("./autosomal_recessive/autosomal_recessive_families.tsv", "r") as fi:
    recessive = "\t".join(fi.read().splitlines()).split("\t")

# Filter out empty string elements in the list
recessiveF = list(filter(None, recessive))

# Get number of appearances of genes 
recessive_keys = list(set(recessiveF))

counts = []

for i in range(0, len(recessive_keys)):
    counts.append(recessiveF.count(recessive_keys[i]))

# Create a dictionary with name of gene and appearances in the prioritization
dict_recessive = dict(zip(recessive_keys, counts))
dict_recessive
```

```python
## AUTOSOMAL RECESSIVE WITH MEDIUM IMPACT

# Read file with autosomal_recessive MEDIUM impact
with open("./autosomal_recessive/autosomal_recessive_MED_families.tsv", "r") as fi:
    recessiveMED = "\t".join(fi.read().splitlines()).split("\t")

# Filter out empty string elements in the list
recessiveMEDF = list(filter(None, recessiveMED))

# Get number of appearances of genes 
recessiveMED_keys = list(set(recessiveMEDF))

counts = []

for i in range(0, len(recessiveMED_keys)):
    counts.append(recessiveMEDF.count(recessiveMED_keys[i]))

# Create a dictionary with name of gene and appearances in the prioritization
dict_recessiveMED = dict(zip(recessiveMED_keys, counts))
dict_recessiveMED
```

```python
## COMPOUND HETEROZYGOTES

# Read file with compount heterozygots
with open("./comp_het/comp_het_families.tsv", "r") as fi:
    comp_het = "\t".join(fi.read().splitlines()).split("\t")

# Filter out empty string elements in the list
comp_hetF = list(filter(None, comp_het))

# Get number of appearances of genes 
comp_het_keys = list(set(comp_hetF))

counts = []

for i in range(0, len(comp_het_keys)):
    counts.append(comp_hetF.count(comp_het_keys[i]))

# Create a dictionary with name of gene and appearances in the prioritization
dict_comp_het = dict(zip(comp_het_keys, counts))
dict_comp_het
```

```python
## GENERAL VARIANTS

# Create a general dictionary with counts of appearances of genes in all tables
all_keys = de_novo_keys + mendel_keys + recessive_keys + recessiveMED_keys + comp_het_keys
general_keys = list(set(all_keys))

counts = []

for i in range(0, len(general_keys)):
    counts.append(all_keys.count(general_keys[i]))

# Create dictionary with counts of all genes
dict_general = dict(zip(general_keys, counts))

df = pd.DataFrame.from_dict(dict_general, orient = 'index')

df.columns = ["counts"]

# Create dataframe with gene counts
df2 = df.sort_values(by='counts', ascending = False).drop("gene")
```

```python
df_common = df2.copy()

df_common2 = df_common[df_common.counts != 1]

df_common2
```

```python
dfg = pd.DataFrame([dict_de_novo, dict_mendel_error, dict_recessive, dict_recessiveMED, dict_comp_het], columns = general_keys)
dfgF = dfg.fillna(0)
dfgF
```

```python
dfgF.loc["MendelNovo"] = (dfgF.loc[0] - dfgF.loc[1]).abs() + dfgF.loc[0]
buena = dfgF.drop([0, 1])
buena.loc["total"] = buena.select_dtypes(pd.np.number).sum()

buena.to_csv("count_genes_probands.tsv", sep = "\t")
```

```python
transposed = buena.transpose().sort_values(by=["total"], ascending = False).drop(["gene"])
transposed.to_csv("count_genes_probands.tsv", sep = "\t")
```
