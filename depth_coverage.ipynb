```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
plt.style.use('seaborn-whitegrid')
sns.set_style('whitegrid')
```

```python
# Read pandas dataframe from the tsv we created
df = pd.read_csv("vlt_samples/vlt_probes_pertype.csv", sep='\t')
df.head()
```

```python
%matplotlib inline
fig = plt.figure(figsize=(20,20))
sns.set(font_scale=1.5, style='whitegrid')
sns.factorplot("depth", "mean", "sample", data=df, size=10, aspect=1, kind="point", legend=False)
plt.xticks(rotation=50)
leg = plt.legend(fontsize = 14, framealpha = 1, frameon = True)
plt.autoscale()
plt.tight_layout()
plt.savefig("vlt_samples/vlt_mean_depth.png")
plt.show()
plt.close(fig)
```
